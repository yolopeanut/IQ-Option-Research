```python
df = df2
df['LINEARREG'] = pd.Series(LINEARREG(df['close'], timeperiod=5))
df['LINEARREG2'] = pd.Series(LINEARREG(df['close'], timeperiod=6))
df = df.dropna()
df['future_reg'] = np.where(df['LINEARREG'] >= df['close'],1 , 0)
df['future_reg2'] = np.where(df['LINEARREG2'] >= df['close'],1 , 0)
df_group = df.groupby(['Future','future_reg', "future_reg2"]).count()
df.head(50)
```

```python
from tqdm import tqdm
temp = pd.DataFrame(columns = ["First Log", "Second Log","Percent"])
for i in tqdm(range(2, 50)):
    for j in range(2, 52):
        if i != j:      
            df = df2
            df['LINEARREG'] = pd.Series(LINEARREG(df['close'], timeperiod=i))
            df['LINEARREG2'] = pd.Series(LINEARREG(df['close'], timeperiod=j))
            df = df.dropna()
            df['future_reg'] = np.where(df['LINEARREG'] >= df['close'],1 , 0)
            df['future_reg2'] = np.where(df['LINEARREG2'] >= df['close'],1 , 0)
            df_group = df.groupby(['Future','future_reg', "future_reg2"]).count()
#             print(df_group.iloc[[0],:])
            correct = df_group.iloc[[0], :].values[0][0] + df_group.iloc[[7], :].values[0][0]
            total = df_group.iloc[[4], :].values[0][0] + df_group.iloc[[3], :].values[0][0] + df_group.iloc[[0], :].values[0][0] + df_group.iloc[[7], :].values[0][0]

            per = correct/total*100 
            to_append = [i,j,per]
            temp.loc[len(temp)] = to_append

        
temp        
```
